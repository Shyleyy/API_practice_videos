```python
import requests
```

```python
# makes a GET request to the API endpoint and stores the response in a variable
response = requests.get('https://randomfox.ca/floof')

# prints out the status code of the response
fox = response.json()
print(response.status_code)
```

```python
# prints out a string of the returned data
print(response.text)
```

```python
# this is going to show us a dictionary of the returned data
print(response.json())
```

```python
# print out a string of the image URL from the returned data
print(fox['image'])
```
